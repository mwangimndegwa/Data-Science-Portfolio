```python
import pandas as pd

df = pd.read_csv("Example-2024.csv", header = None)
```

```python
df.shape
```

```python
df.columns = ['Title', 'Artist', 'Date', 'Genre', 'Language', 'Lyrics', 'URL']

df.head()
```

```python
df.info()
```

```python
print(df['Artist'].nunique())
print(df['Language'].nunique())
```

```python
df.head()
```

```python
df['Artist'].value_counts()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

print(df['Date'].min())
print(df['Date'].max())
```

```python
df['Month'] = df['Date'].dt.month

df.info()
```

```python
df['Month'].value_counts()
```

```python
df['Lyrics_Length'] = df['Lyrics'].str.len()
df['Lyrics_Length'].max()
```

```python
df['Word_Count'] = df['Lyrics'].str.split().str.len()

df['Word_Count'].mean().round(0)
```

```python
df['Genre'].str.contains('pop', case = False).sum()
```

```python
df[df['Word_Count'] < 10]
```

```python
df['Lyrics'] = df['Lyrics'].str.replace('\n', ' ', regex=True)
```

```python
df.groupby('Artist')['Word_Count'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('Genre')['Word_Count'].mean().sort_values(ascending=False).head()
```

```python
df['Month'].value_counts().sort_index(ascending=True).plot(kind='bar')
```

```python
import re

df['Lyrics'].str.count('love', flags=re.IGNORECASE).sum()
```
